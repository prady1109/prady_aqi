```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Importing the dataset
dataset = pd.read_csv('rkpur.csv')
pd.set_option("max_rows", None)
```

```python
dataset.shape
```

```python
dataset['AT']=dataset['Temp'].where((dataset['AT'].isnull() | dataset['AT'].isna()),dataset['AT'])
```

```python
dataset['From Date']=pd.to_datetime(dataset['From Date']) 
dataset['Month'] = dataset['From Date'].dt.month
```

```python
dt = dataset.loc[:,['PM2.5']].values
dt=pd.DataFrame(dt).interpolate().values
X = []
y= []
for i in range(15, 1627):
    X.append(dt[i-15:i, 0])
    y.append(dt[i, 0])
X, y = np.array(X), np.array(y)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#feature scaling

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train[:,11:] = sc_X.fit_transform(X_train[:,11:])
X_test[:,11:]= sc_X.transform(X_test[:,11:])
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train)
y_test=sc_y.transform(y_test)

```

```python
#implementing model
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.wrappers.scikit_learn import KerasRegressor
from keras import losses
```

```python
def buildrnn(unit=32):
    regressor = Sequential()
    regressor.add(LSTM(units = unit, return_sequences = True, input_shape = (X_train.shape[1], 1)))
    regressor.add(Dropout(0.2))
    regressor.add(LSTM(units = unit, return_sequences = True))
    regressor.add(Dropout(0.2))
    regressor.add(LSTM(units = unit))
    regressor.add(Dropout(0.2))
    regressor.add(Dense(units = 1))
    regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
    return regressor
```

```python
regressor= KerasRegressor(build_fn = buildrnn,epochs=500,batch_size=32)

results=regressor.fit(X_train,y_train)
regressor.score(X_test,y_test)
```

```python
regressor.score(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
from sklearn import metrics
pm25test=metrics.mean_squared_error(y_test,y_pred)
print((pm25test))
```

```python
regressor= KerasRegressor(build_fn = buildrnn,epochs=500,batch_size=32)
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = regressor, X = X, y = y, cv = 5,n_jobs=-1)
print(accuracies.mean(),accuracies.std())
```
